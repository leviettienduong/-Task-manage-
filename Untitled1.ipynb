```python
gpa = float(input("Nhập GPA: "))

if gpa >= 8.5:
    print("Xuất sắc")
elif gpa >= 7.0:
    print("Giỏi")
elif gpa >= 5.5:
    print("Khá")
elif gpa >= 4.0:
    print("Trung bình")
else:
    print("Yếu")
```

```python
n = int(input("Nhập n: "))
tong = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        tong += i

print("Tổng các số chẵn:", tong)
```

```python
matkhau_dung = "python123"
dem = 0

while dem < 3:
    matkhau = input("Nhập mật khẩu: ")

    if matkhau == matkhau_dung:
        print("Đăng nhập thành công")
        break
    else:
        dem += 1

if dem == 3:
    print("Tài khoản bị khóa")
```
