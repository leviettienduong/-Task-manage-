```python
n = int(input("Nhap n: "))

tong = 0

for i in range(n):
    x = float(input(f"Nhap so thu {i+1}: "))
    tong += x

trung_binh = tong / n

print("Trung binh cong =", trung_binh)
```

```python
n = int(input("Nhap n: "))

dem = 0

for i in range(n):
    x = int(input(f"Nhap so thu {i+1}: "))
    if x % 2 == 0:
        dem += 1

print("So luong so chan =", dem)
```

```python
n = int(input("Nhap n: "))

x = float(input("Nhap so thu 1: "))
max_val = x
min_val = x

for i in range(2, n + 1):
    x = float(input(f"Nhap so thu {i}: "))
    if x > max_val:
        max_val = x
    if x < min_val:
        min_val = x

print("Gia tri lon nhat =", max_val)
print("Gia tri nho nhat =", min_val)
```

```python
a = []
for i in range(5):
    a.append(float(input(f"Nhap so thu {i+1}: ")))

for i in range(4):
    for j in range(i + 1, 5):
        if a[i] > a[j]:
            a[i], a[j] = a[j], a[i]

print("Day tang dan:", a)
```

```python
def giaithua(n):
    if n == 0:
        return 1
    return n * giaithua(n - 1)

n = int(input("Nhap n: "))
print("Giai thua =", giaithua(n))
```
