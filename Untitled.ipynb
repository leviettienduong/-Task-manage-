```python
nums = list(map(int, input("Nhập 10 số nguyên, cách nhau bởi khoảng trắng: ").split()))

neg = pos = zero = 0

for x in nums:
    if x < 0:
        neg += 1
    elif x > 0:
        pos += 1
    else:
        zero += 1

print("Số âm:", neg)
print("Số dương:", pos)
print("Số bằng 0:", zero)
```

```python
n = int(input("Nhập số n: "))

if n < 2:
    print("Không phải số nguyên tố")
else:
    is_prime = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Là số nguyên tố")
    else:
        print("Không phải số nguyên tố")
```
